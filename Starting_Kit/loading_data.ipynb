```python
import numpy as np
import os
import sklearn as sk
```

```python
x_path = "trainValTest_N=2197469_prod2_features_model=efficientnetB3_tsfrm=dataset_labeledOnly.npy"
# or .npz if you did not extract it

y_path = "y.npy"




X = np.load(x_path)
y = np.load(y_path)

# Getting rid of images with no bugs.
X_bug = X[y != 0]
y_bug = y[y != 0]


print("X shape :",X_bug.shape)
print("Y shape : ",y_bug.shape)

X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X_bug, y_bug, test_size=0.33)

```

```python
counts, vals = np.unique(y_bug, return_counts=True)

for c, occ in zip(counts, vals):
    print(f"Label {c} : {occ} occurences")
```

```python
model = sk.linear_model.SGDClassifier(
    loss="log_loss",
    max_iter=10,
    n_jobs=1
)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

print(sk.metrics.confusion_matrix(y_test, y_pred))


trainScore = model.score(X_train, y_train)
testScore = model.score(X_test, y_test)
print(f"Train Score : {trainScore}")
print(f"Test Score : {testScore}")
```
